```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 100, 1000) # start, finish, n points
```

```python
power_mark_1 = 1 * t
```

```python
power_makr_2 = 4 * (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Solar power output mark-1 vs. mark-2')
plt.xlabel('time (in days)')
plt.ylabel('power (in kj)')
ax.set_xlim([0, 100])
ax.set_ylim([0, 100])
ax.plot(t, power_mark_1, c='green')
ax.plot(t, power_makr_2, c='orange')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
